```python
# objectiv2   [ check water usage]
import pandas as pd
import matplotlib.pyplot as plt

# Load the verdant data
df = pd.read_excel(r"C:\Users\devas\OneDrive\Desktop\surfacelift_rajasthan.xlsx", sheet_name=0)

# Calculate total water used per season per row
df['water_kharif'] = df['avg_pumping_per_day_kharif_season'] * df['pump_operating_days_kharif_season']
df['water_rabi'] = df['avg_pumping_per_day_rabi_season'] * df['pump_operating_days_rabi_season']
df['water_perennial'] = df['avg_pumping_per_day_perennial_season'] * df['pump_operating_days_perennial_season']
df['water_other'] = df['avg_pumping_per_day_other_season'] * df['pump_operating_days_other_season']

# Sum to get total water usage per scheme
df['total_water_used'] = df[['water_kharif', 'water_rabi', 'water_perennial', 'water_other']].sum(axis=1)

# Group by district and sum total water usage
district_water = df.groupby('district_name')['total_water_used'].sum().sort_values(ascending=False)

# Print overall total to celebrate the insight!
print(" Total Estimated Water Used (All Districts):", district_water.sum())

# Plot the captivating chart
plt.figure(figsize=(14, 6))
district_water.plot(kind='bar', color='seagreen')
plt.title('Estimated Total Water Usage by District')
plt.xlabel('District')
plt.ylabel('Total Estimated Water Used (unit-days)')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# obj2 ka [by pie chart]

import matplotlib.pyplot as plt

# Top 5 districts
top5 = district_water.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set2.colors)
plt.title('Top 5 Districts by Water Usage (Pie Chart)')
plt.show()
```

```python
#  obj2 [ by horizontal bar]
top10 = district_water.head(10)

plt.figure(figsize=(10, 6))
top10.plot(kind='barh', color='coral')
plt.xlabel('Estimated Water Used')
plt.ylabel('District')
plt.title('Top 10 Districts by Water Usage (Horizontal Bar)')
plt.gca().invert_yaxis()  # To show highest on top
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Sort alphabetically
sorted_district = district_water.sort_index()

plt.figure(figsize=(14, 5))
plt.plot(sorted_district.index, sorted_district.values, marker='o', color='teal')
plt.title('Estimated Water Usage by District (Line Chart)')
plt.xlabel('District')
plt.ylabel('Water Used')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.fill_between(sorted_district.index, sorted_district.values, color='skyblue', alpha=0.5)
plt.plot(sorted_district.index, sorted_district.values, color='blue')
plt.title('Area Chart: Water Usage by District')
plt.xlabel('District')
plt.ylabel('Estimated Water Used')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#objec5
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel(r"C:\Users\devas\OneDrive\Desktop\surfacelift_rajasthan.xlsx")

# Count of water distribution methods
distribution_counts = df['water_distribution_method_name'].value_counts()

# Prepare data for dot graph
methods = distribution_counts.index.tolist()
counts = distribution_counts.values
y_positions = list(range(len(methods)))

# Create dot graph
plt.figure(figsize=(10, 6))
plt.scatter(counts, y_positions, color='darkorange', s=100)

# Add labels
for i, count in enumerate(counts):
    plt.text(count + 1, y_positions[i], str(count), va='center')

plt.yticks(y_positions, methods)
plt.xlabel('Number of Schemes')
plt.title('Dot Graph of Water Distribution Methods')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#objec5
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_excel(r"C:\Users\devas\OneDrive\Desktop\surfacelift_rajasthan.xlsx")

# Set plot style
sns.set(style="whitegrid")

# Plot using Seaborn stripplot
plt.figure(figsize=(10, 6))
sns.stripplot(
    y='water_distribution_method_name',
    data=df,
    size=8,
    color='teal',
    jitter=True  # Adds horizontal spread so dots don't overlap
)

plt.title('Dot Plot of Water Distribution Methods')
plt.xlabel('Frequency of Schemes')
plt.ylabel('Water Distribution Method')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel(r"C:\Users\devas\OneDrive\Desktop\surfacelift_rajasthan.xlsx")

# Select numerical features for correlation
cols = [
    'scheme_construction_cost',
    'scheme_machinery_cost',
    'scheme_maintainence_cost',
    'horse_power_of_lifting_device',
    'ipc_total',
    'ipu_total'
]

# Drop rows with NaNs in selected columns
df_corr = df[cols].dropna()

# Compute correlation matrix
correlation_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Numerical Features Correlation")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
